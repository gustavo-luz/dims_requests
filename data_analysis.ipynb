```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datasets/unfiltered_df.csv')
print(df.columns)
df.groupby(['Date']).size()
```

```python
df["Date"] = df["Date"].astype("datetime64")


df.groupby([df["Date"].dt.month,df["Date"].dt.day]).count().plot(kind="bar")
```

```python
df.hist(column='Date',bins=40)
plt.xticks(rotation='vertical')

plt.show()
```

```python
df_15_25 = df[(df['Date'] >= '2022-06-15') & (df['Date'] <= '2022-06-25')]
df_15_25.hist(column='Date',bins=40)#title='Amount of data sent at test'
print(df_15_25.groupby(['Date']).size().sum())
print(df_15_25.groupby(['Date']).size())
print(df_15_25.groupby(['Date']).size().mean())
```
